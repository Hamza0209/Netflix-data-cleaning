```python
import pandas as pd

#read the file
df = pd.read_csv("netflix_full.csv")
df.head()
```

Check basic info, missing values, and duplicates

```python
print("ðŸ“Œ Dataset Info:")
print(df.info())

print("\nðŸ“Œ Missing Values:")
print(df.isnull().sum())

print("\nðŸ“Œ Duplicate Rows:", df.duplicated().sum())
```

Remove Duplicate Rows

```python
# Check duplicates before removing
print("Duplicates Before:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Check again
print("Duplicates After:", df.duplicated().sum())
```

Handle Missing Values

```python
# Fill missing numerical values with mean
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col] = df[col].fillna(df[col].mean())

# Fill missing text values with mode
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Standardize Text Formatting

```python
# Convert text to lowercase and remove extra spaces
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.lower().str.strip()

print("âœ” Text formatting standardized.")
```

Fix Date Format (Netflix dataset has release_year and date_added)

```python
# Convert date_added column to proper datetime
if 'date_added' in df.columns:
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

print(df.head())
```

Clean Column Names

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")

print("âœ” Column names cleaned:")
df.columns
```

Export the Cleaned File

```python
df.to_csv("cleaned_netflix_data.csv", index=False)
print("ðŸŽ‰ Cleaned file saved as cleaned_netflix_data.csv!")
```
